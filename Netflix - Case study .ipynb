* Values in **Duration** column has different units, depending upon the type of content. For "TV shows" its given in seasons and for "Movies its given in minutes. And the datatype of that column is "object"

**Filling Null values**

```python
# Filling nan Values with "Unknown" in director column
c = df["director"].apply(lambda x: str(x).split(", ")).tolist()
df1 = pd.DataFrame(c, index=df["title"])
df1 = df1.stack().reset_index()
df1 = df1[["title",0]]
df1.columns= ["title", "director"]
df1["director"].replace(to_replace="nan", value="Unknown", inplace= True)
df1["director"].isnull().sum()
```

```python
#stacking genre column with title

df.head()
d = df["listed_in"].apply(lambda x: str(x).split(", ")).tolist()
df2 = pd.DataFrame(d, index=df["title"])
df2 = df2.stack().reset_index()
df2 = df2[["title",0]]
df2.columns = ["title","listed_in"]
df2["listed_in"].isnull().sum()
```

```python
#stacking country columns with title

e = df["country"].apply(lambda x: str(x).split(", ")).tolist()
df3 = pd.DataFrame(e, index=df["title"])
df3=df3.stack().reset_index()
df3=df3[["title",0]]
df3.columns = ["title","country"]
df3["country"].replace(to_replace= "nan", value = "Unknown", inplace = True)
```

```python
#Splitting and stacking cast along with title

f = df["cast"].apply(lambda x: str(x).split(", ")).tolist()
df_cast = pd.DataFrame(f, index=df["title"])
df_cast = df_cast.stack()
df_cast= pd.DataFrame(df_cast)

df_cast.reset_index(inplace = True)
df_cast=df_cast[["title",0]]
df_cast.columns= ["title", "cast"]
df_cast["cast"].replace(to_replace = "nan", value = "Unknown", inplace = True)
df_cast["cast"].isnull().sum()
```

```python
#Splitting the column duration and slicing only the integers

du = df["duration"].str.split(" ").str[0].tolist()
dura = pd.DataFrame(du, index= df["title"])
dura = dura.stack().reset_index()
dura = dura[["title",0]]
dura.columns= ["title", "duration"]
dura["duration"].replace(to_replace="nan", value= 0, inplace= True)
```

**Converting date to datetime format and extracting years, month and date from it**

```python
df["date_added"] = pd.to_datetime(df["date_added"])
```

```python
df["df_year_added"] = df["date_added"].dt.year
df["df_month_added"] = df["date_added"].dt.month
df["df_month_name_added"] = df['date_added'].dt.month_name()
```
